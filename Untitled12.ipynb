```python
##speacial methods in python
class Book():
    def __init__(self,title,author,pages):
        self.title = title
        self.author = author
        self.pages = pages
    def __str__(self):
        return f"{self.title} by {self.author}"
    def __len__(self):
        return self.pages
    def __del__(self):
        print("the Book has been deleted")
        
```

```python
mybook = Book("unwritten India","Anjuna",200)
```

```python
print(mybook)
```

```python
print(len(mybook))
del(mybook)
```
