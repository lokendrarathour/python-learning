```python
def ask_int():
    while True:
        try:
            res = int(input("Please enter the number:"))
            print(res)
        except:
            print("The number entered by you is not a integer")
            continue
        else:
            print("Number entered by you is correct")
            break
        finally:
            print("this is the final statement")
            
ask_int()
```
