```python
class Geometry():
    def __init__(self,side=10):
        ## attribute override 
        self.side = side
        self.geomFlag = 1
        print("This is the Space of Geometry !!")
    def circum(self):
        return(4*4)    
    
```

```python
class Maths():
    def __init__(self,side=10):
        ## attribute override 
        self.side = side
        self.geomFlag = 1
        print("This is the Space of MathS !!")
    def circum(self):
        return(4*4)
```

```python
class Circle(Geometry,Maths):
    ## Class Object Atribute 
    pi = 3.14
    def __init__(self,radius=1):
        self.radius = radius
        self.area = self.radius*self.radius*2
        self.circumference = self.radius*2*Circle.pi
         ## this is how u instantiate a class
        Geometry.__init__(self,radius)
        Maths.__init__(self,radius)
       
    def diameter(self):
        return(self.radius*2)
```

```python
class Square(Geometry): 
    def __init__(self,side=1):
        self.side = side
        self.area = self.side*self.side
         ## this is how u instantiate a class
        Geometry.__init__(self,side)

    def circum(self):
        return(self.side*4)
```

```python
mysquare = Square()
mycircle = Circle()
```
