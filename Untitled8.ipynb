```python
class Circle():
    '''
    This is the code for Circle
    Function includes : 
        radius
        area
        Circumferance
    '''
    ## class object attribute
    pi = 3.14
    ## constructor  ## with argument 
    def __init__(self,radi=1):
        self.radi = radi
        # area is an type of attribute which is self defined 
        self.area = self.pi*self.radi*self.radi
    ## methods 1 with argument 
    def circumference(self,radi):
        circum = 2*self.pi*self.radi
        return(circum)
    
    ## method 2 without arguments 
    def diameter(self):
        return(2*self.radi)
        
```
