```python
class Dog():
    #pass
    ## init is the construct which is automatically called when the fuction
    ## is called
    ## self is an instance of the object 
    ## breed here in def definition is positional argument
    ## breed in the body of def is something that is called upon
    ## class object attribute
    typeAnimal = 'mammal'
    
    def __init__(self,mybreed,name,dogtype,spots):
        self.breed = mybreed
        self.name = name
        self.dogtype = dogtype
        self.spots = spots
        print(type(self.spots))
    
    def bark(self,number):
        print("Dog Bark  {} and the number is {}".format(self.name,number))

myDog = Dog(mybreed='lab',name="sandy",dogtype="pet",spots=True)
print(type(myDog))
print(myDog.breed)
print(myDog.spots)
print(myDog.typeAnimal)
myDog.bark(10)
```

```python
class Circle():
    ## Atribute oject
    pi = 3.14
    def __init__(self,radius=1):
        self.radius = radius
        self.area = self.radius*self.radius*2
        #self.circumference = self.radius*2*self.pi
        self.circumference = self.radius*2*Circle.pi
    def diameter(self):
        return(self.radius*2)
        
```

```python
my_circle = Circle()
print(my_circle.pi)
print(my_circle.circumference)
print("diameter of the circle is {}".format(my_circle.diameter()))
```
