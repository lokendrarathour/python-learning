```python
#def 
def readfile(filename):
    mfh = open(filename,'r')
    return mfh

def process(mfhandle):
    ml = mfhandle.readlines()
    # counting the number of line
    count = len(ml)
    for line in ml:
        #print(line)
        # logical code can be added
        singleline = line.split()
        print(singleline)
        print(singleline)
    return count
    
    
# calling and working 

mfhandle = readfile("gablearning.txt")
filec = process(mfhandle)
print(filec)


```
