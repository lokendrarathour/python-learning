```python
class Human():
    def __init__(self,name):
        self.name = name        
    def speak(self):
        return("Hahahahahaa")
```

```python
class Dog():
    def __init__(self,name):
        self.name = name
    def speak(self):
        return("woaaf woaf")
```

```python
class Cat():
    def __init__(self,name):
        self.name = name
    def speak(self):
        return("Meaau")
```

```python
myhum = Human("Joh")
myDog = Dog("Sandy")
myCat = Cat("Spiky")
```

```python
print("Hey this is {} and I speak {}".format(myhum.name,myhum.speak()))
print("Hey this is {} and I speak {}".format(myDog.name,myDog.speak()))
print("Hey this is {} and I speak {}".format(myCat.name,myCat.speak()))
```

```python
for i in [myhum,myDog,myCat]:
    print(type(i))
    print(i.speak())
    print(type(i.speak()))

```
